```python
def overview(data):
    '''
    Erstelle einen Überblick über einige Eigenschaften der Spalten eines DataFrames.
    VARs
        data: Der zu betrachtende DataFrame
    RETURNS:
        None
    '''
    if data.shape[0] == 0:
        raise ValueError("DataFrame has no rows.")
        
    display(pd.DataFrame({
        'dtype': data.dtypes,
        'total': data.count(),
        'missing': data.isna().sum(),
        'missing%': data.isna().mean() * 100,
        'n_uniques': data.nunique(),
        'uniques%': data.nunique() / data.shape[0] * 100,
        'uniques': [data[col].unique() for col in data.columns]
    }))

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set theme and style for a clean modern look
sns.set_theme(style="whitegrid")

# Update color palette with stronger blue
palette = {"0.0": "#1f77b4", "1.0": "#ffb347", "2.0": "#ff6961"}  # Strong blue, orange, red

# Ensure 'Diabetes_012' column is of type string
df['Diabetes_012'] = df['Diabetes_012'].astype(str)

# Create the plot with enhancements
plt.figure(figsize=(8, 4))  # Adjust figure size
sns.histplot(data=df, x='BMI', hue='Diabetes_012', palette=palette, kde=True, binwidth=2)

# Add title and customize axis labels
plt.title("BMI Distribution by Diabetes Status", fontsize=16, fontweight='bold')
plt.xlabel("Body Mass Index (BMI)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Set x-axis limit to a range from 0 to 60
plt.xlim(0, 60)

# Customize ticks and legend placement
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Diabetes Status", loc="upper right", fontsize=10)

# Tighten layout for better aesthetics
plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="whitegrid")

# Define a stronger color palette
palette = {"0.0": "#1f77b4", "1.0": "#2ca02c", "2.0": "#ff7f0e"}  # Blue, Green, and Orange

# Ensure 'Diabetes_012' column is of type string
df['Diabetes_012'] = df['Diabetes_012'].astype(str)

# Create subplots for the four variables with improved layout
fig, axes = plt.subplots(2, 2, figsize=(14, 12))  # Larger figure size for clarity

# Stroke Plot
sns.histplot(data=df, x='Stroke', hue='Diabetes_012', palette=palette, kde=True, binwidth=0.5, ax=axes[0, 0])
axes[0, 0].set_title("Stroke Distribution by Diabetes Status", fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel("Stroke (Binary)", fontsize=12)
axes[0, 0].set_ylabel("Frequency", fontsize=12)

# PhysHlth Plot
sns.histplot(data=df, x='PhysHlth', hue='Diabetes_012', palette=palette, kde=True, binwidth=2, ax=axes[0, 1])
axes[0, 1].set_title("Physical Health Distribution by Diabetes Status", fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel("Number of Poor Physical Health Days", fontsize=12)
axes[0, 1].set_ylabel("Frequency", fontsize=12)

# HeartDiseaseorAttack Plot
sns.histplot(data=df, x='HeartDiseaseorAttack', hue='Diabetes_012', palette=palette, kde=True, binwidth=0.5, ax=axes[1, 0])
axes[1, 0].set_title("Heart Disease/Attack Distribution", fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel("Heart Disease or Attack (Binary)", fontsize=12)
axes[1, 0].set_ylabel("Frequency", fontsize=12)

# Age Plot
sns.histplot(data=df, x='Age', hue='Diabetes_012', palette=palette, kde=True, binwidth=5, ax=axes[1, 1])
axes[1, 1].set_title("Age Distribution by Diabetes Status", fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel("Age (Years)", fontsize=12)
axes[1, 1].set_ylabel("Frequency", fontsize=12)

# Adjust layout to avoid overlaps and improve clarity
plt.tight_layout()
plt.show()

```
