```python
import pandas as pd 
#df = pd.read_csv(r"C:\Users\luisz\Desktop\diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
df = pd.read_csv(r"C:\Users\luisz\Desktop\GASTRIC-CANCER2\diabetes_dataset.csv")
#df['Diabetes_Binary'] = df['Diabetes_012']
df.rename(columns={"Diabetes_012": "Diabetes_Binary"}, inplace=True)
#df['Diabetes_012'] = df['Diabetes_Binary']
#df = df[df["Diabetes_Binary"] != 1.0]
#df["Diabetes_Binary"] = df["Diabetes_Binary"].replace(2.0, 1.0)
df["Diabetes_Binary"] = df["Diabetes_Binary"].apply(lambda x: 1 if x in [1, 2] else 0)

print(df)
```

```python
def overview(data):
    '''
    Erstelle einen Überblick über einige Eigenschaften der Spalten eines DataFrames.
    VARs
        data: Der zu betrachtende DataFrame
    RETURNS:
        None
    '''
    if data.shape[0] == 0:
        raise ValueError("DataFrame has no rows.")
        
    display(pd.DataFrame({
        'dtype': data.dtypes,
        'total': data.count(),
        'missing': data.isna().sum(),
        'missing%': data.isna().mean() * 100,
        'n_uniques': data.nunique(),
        'uniques%': data.nunique() / data.shape[0] * 100,
        'uniques': [data[col].unique() for col in data.columns]
    }))

```

```python
df['Overweight'] = df['BMI'].apply(lambda x: 1 if x >= 25 else 0)
```

```python
import numpy as np

# 🔥 Interaction Features
df['BMI_Age'] = df['BMI'] * df['Age']  # How BMI interacts with aging
df['BMI_Exercise'] = df['BMI'] * df['PhysActivity']  # Effect of exercise on high BMI

# 🔍 Health Condition Flags
df['ObesityRisk'] = df['BMI'].apply(lambda x: 1 if x >= 30 else 0)  # Stronger obesity indicator
df['HighBP_HighChol'] = df['HighBP'] * df['HighChol']  # Comorbid risk factor

# 🏃 Behavioral Features
df['HealthyHabits'] = df[['PhysActivity', 'Fruits', 'CholCheck']].sum(axis=1)  # Count of preventive actions
df['RiskFactorScore'] = df[['HighBP', 'HighChol', 'Smoker', 'PhysHlth']].sum(axis=1)  # Total risk aggregation

# 💰 Socioeconomic Impact
df['LowIncome_HealthAccess'] = df['Income'] * df['AnyHealthcare']  # Measures healthcare limitations
df['Education_HealthAwareness'] = df['Education'] * df['CholCheck']  # Impact of education on medical awareness

# 🔧 Scale Continuous Features (Standardization for smooth learning)
continuous_features = ['BMI', 'Age', 'Income', 'BMI_Age', 'BMI_Exercise']
df[continuous_features] = (df[continuous_features] - df[continuous_features].mean()) / df[continuous_features].std()

# 📊 Final Feature Set
selected_features = ['GenHlth', 'HighBP', 'HighChol', 'PhysActivity', 'BMI', 'Age', 'ObesityRisk', 
                     'Fruits', 'DiffWalk', 'Education', 'HeartDiseaseorAttack', 'PhysHlth', 'Smoker', 
                     'BMI_Age', 'BMI_Exercise', 'HealthyHabits', 'RiskFactorScore', 
                     'LowIncome_HealthAccess', 'Education_HealthAwareness']


```

```python
Target = 'Diabetes_Binary'
```

```python
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
```

```python
from sklearn.model_selection import train_test_split
X = df[Selected_features]
y = df[Target]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , test_size=0.2)
```

```python
cat  = CatBoostClassifier(verbose=0)
lgb = LGBMClassifier(verbose = -1)
xgb = XGBClassifier()
```

```python
model = [('cat', cat), ('lgb', lgb), ('xgb', xgb)]

clf = VotingClassifier(model , voting='soft' , n_jobs = -1)
```

```python
clf.fit(X_train , y_train)
```

```python
y_probs = clf.predict_proba(X_test)
```

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

y_probs_positive_class = y_probs[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_probs_positive_class)

auprc = auc(recall, precision)

plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AUC = {auprc:.2f})')
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, auc, accuracy_score, classification_report
import matplotlib.pyplot as plt

# 🔍 Predictions & Probability Extraction
y_probs_positive_class = y_probs[:, 1]  # Select probability for positive class
y_pred = (y_probs_positive_class >= 0.72).astype(int)  # Adjust threshold if needed

# 📊 Compute Classification Metrics
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, output_dict=True)

print(report.keys())  # See what class labels exist in the classification report


precision = report["1.0"]["precision"]
recall = report["1.0"]["recall"]
f1_score = report["1.0"]["f1-score"]


# 🔥 Precision-Recall Curve Calculation
precisions, recalls, _ = precision_recall_curve(y_test, y_probs_positive_class)
auprc = auc(recalls, precisions)

# 🏆 Display Model Performance
print(f"✅ Adjusted Accuracy: {accuracy:.4f}")
print(f"📌 Precision (Class 1): {precision:.4f}")
print(f"📌 Recall (Class 1): {recall:.4f}")
print(f"📌 F1-score (Class 1): {f1_score:.4f}")
print(f"📈 Precision-Recall AUC: {auprc:.4f}")

# 🔍 Plot Precision-Recall Curve
plt.figure(figsize=(8,6))
plt.plot(recalls, precisions, marker='.', label=f'PR-AUC={auprc:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()

```

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_recall_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 🔥 Step 1: Fine-Tune Class Weighting for Better Recall
logistic_model = LogisticRegression(class_weight={0:1, 1:3.5}, C=0.14, penalty='l1', solver='saga', max_iter=300)

# 🚀 Step 2: Adjust XGBoost Model for False Negative Reduction
xgb_model = XGBClassifier(n_estimators=300, max_depth=5, learning_rate=0.1, subsample=0.8, colsample_bytree=0.8, scale_pos_weight=3.5)

# 🏆 Step 3: Ensemble Learning (Logistic Regression + XGBoost)
ensemble_model = VotingClassifier(estimators=[('lr', logistic_model), ('xgb', xgb_model)], voting='soft')

# 🔍 Step 4: Train the Hybrid Model
ensemble_model.fit(X_train, y_train)

# 📊 Step 5: Make Predictions with Adjusted Threshold
y_prob = ensemble_model.predict_proba(X_test)[:, 1]
threshold = 0.67  # Lowering threshold to catch more true positives
y_pred = (y_prob >= threshold).astype(int)

# 🔥 Step 6: Evaluate Model Performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, output_dict=True)

# ✅ Fix: Dynamic Key Access for Classification Report
precision = report.get("1", {}).get("precision", None) or report.get("1.0", {}).get("precision", None)
recall = report.get("1", {}).get("recall", None) or report.get("1.0", {}).get("recall", None)
f1_score = report.get("1", {}).get("f1-score", None) or report.get("1.0", {}).get("f1-score", None)

# 📈 Precision-Recall Curve AUC Calculation
precisions, recalls, _ = precision_recall_curve(y_test, y_prob)
auprc = auc(recalls, precisions)

# 📊 Step 7: Print Performance Metrics
print(f"✅ Adjusted Accuracy: {accuracy:.4f}")
if precision is not None and recall is not None and f1_score is not None:
    print(f"📌 Precision (Class 1): {precision:.4f}")
    print(f"📌 Recall (Class 1): {recall:.4f}")
    print(f"📌 F1-score (Class 1): {f1_score:.4f}")
else:
    print("⚠️ Warning: Class 1 metrics missing, check model predictions!")
print(f"📈 Precision-Recall AUC: {auprc:.4f}")

# 🔄 Step 8: Plot Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

# 🔍 Step 9: Plot Precision-Recall Curve
plt.figure(figsize=(8,6))
plt.plot(recalls, precisions, marker='.', label=f'PR-AUC={auprc:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```
