```python
def overview(data):
    '''
    Erstelle einen Überblick über einige Eigenschaften der Spalten eines DataFrames.
    VARs
        data: Der zu betrachtende DataFrame
    RETURNS:
        None
    '''
    if data.shape[0] == 0:
        raise ValueError("DataFrame has no rows.")
        
    display(pd.DataFrame({
        'dtype': data.dtypes,
        'total': data.count(),
        'missing': data.isna().sum(),
        'missing%': data.isna().mean() * 100,
        'n_uniques': data.nunique(),
        'uniques%': data.nunique() / data.shape[0] * 100,
        'uniques': [data[col].unique() for col in data.columns]
    }))

```

```python
df['Overweight'] = df['BMI'].apply(lambda x: 1 if x >= 25 else 0)  
print(df.columns)  
```

```python
# Make a copy of df before applying modifications
df_mod = df.copy()

# Convert all individuals with Diabetes_012 == 2 to 1
df_mod.loc[df_mod['Diabetes_012'] == 2, 'Diabetes_012'] = 1

# Convert Diabetes_012 == 0 to 1 where all four risk factors are 1
df_mod.loc[(df_mod['Diabetes_012'] == 0) & 
                (df_mod[['Overweight', 'HighBP', 'HighChol']].eq(1).all(axis=1)), 'Diabetes_012'] = 1

# Verify the updated counts
counts_updated = df_mod['Diabetes_012'].value_counts()
print("Updated count of 0s and 1s in Diabetes_012 column:")
print(counts_updated)
```

```python

import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

df_features = df_mod.drop(columns=['Diabetes_012'])
target = df_mod['Diabetes_012']  # Target variable

X_train, X_test, y_train, y_test = train_test_split(df_features, target, test_size=0.2, random_state=42)

# Convert X_train to DataFrame to ensure column names are retained
X_train_df = pd.DataFrame(X_train, columns=df_features.columns)

# Train Logistic Regression Model
logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(X_train_df, y_train)

# Extract Feature Importance for ALL 31 Features
logreg_importance = pd.Series(logreg.coef_[0], index=X_train_df.columns)

# Print Top Features
print("Top Features (Logistic Regression):")
print(logreg_importance.sort_values(ascending=False).head(10))
import matplotlib.pyplot as plt

# Sort features by importance
logreg_importance_sorted = logreg_importance.sort_values(ascending=False)

# Plot feature importance as a bar chart
plt.figure(figsize=(10, 6))
logreg_importance_sorted.head(10).plot(kind='bar', color='skyblue')

# Formatting the plot
plt.title("Top 10 Features (Logistic Regression)")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.xticks(rotation=45) 

# Show the plot
plt.show()

import os
import matplotlib.pyplot as plt

# ✅ Define the correct save path inside notebooks directory
save_path = "C:/Users/luisz/Desktop/work/projects/Dist/DIABETES/notebooks/confusion_matrix.png"

# ✅ Ensure directory exists before saving
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# ✅ Save the figure with high quality
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"✅ Die Abbildung wurde erfolgreich gespeichert unter: {save_path}")


```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Define feature names & convert data to DataFrame format (only once)
feature_names = df_mod.drop(columns=['Diabetes_012']).columns  
X_train_df = pd.DataFrame(X_train, columns=feature_names)
X_test_df = pd.DataFrame(X_test, columns=feature_names)

# Prediction using VotingClassifier
y_probs = clf.predict_proba(X_test_df)

# Train Logistic Regression model
model = LogisticRegression()
model.fit(X_train_df, y_train)  
y_pred = model.predict(X_test_df)

```

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt

y_probs_positive_class = y_probs[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_probs_positive_class)

auprc = auc(recall, precision)

plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AUC = {auprc:.2f})')
plt.show()
```

```python
from sklearn.metrics import classification_report

# Generate classification report
report = classification_report(y_test, y_pred, target_names=["No Diabetes (0)", "Diabetes/Risk (1)"])

# Print full performance report
print("Performance Metrics:")
print(report)
```

```python

import pandas as pd
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import precision_recall_curve, classification_report, auc
import matplotlib.pyplot as plt
import numpy as np

# Define selected feature names explicitly
selected_features = [
    'HighChol', 'HighBP', 'Overweight', 'CholCheck', 'GenHlth',
    'HeartDiseaseorAttack', 'Sex', 'Age', 'DiffWalk', 'Stroke'
]

# Convert X_train and X_test to DataFrames NOTE: X_train X_test
X_train_df = pd.DataFrame(X_train_selected, columns=selected_features)
X_test_df = pd.DataFrame(X_test_selected, columns=selected_features)

# Ensure only valid features are selected
valid_features = [feat for feat in selected_features if feat in X_train_df.columns]

X_train_selected = X_train_df[valid_features]
X_test_selected = X_test_df[valid_features]

# Print final feature set to verify
print("Final features used:", X_train_selected.columns.tolist())


# Train Logistic Regression model using selected features
logreg = LogisticRegression()
logreg.fit(X_train_selected, y_train)  
y_pred_logreg = logreg.predict(X_test_selected)

# Initialize VotingClassifier with selected models
cat = CatBoostClassifier(verbose=0)
lgb = LGBMClassifier(verbose=-1)
xgb = XGBClassifier()

model_list = [('cat', cat), ('lgb', lgb), ('xgb', xgb)]
clf = VotingClassifier(model_list, voting='soft', n_jobs=-1)

# Fit VotingClassifier using selected features
clf.fit(X_train_selected, y_train)

# Prediction using VotingClassifier
y_probs = clf.predict_proba(X_test_selected)

# Adjust threshold (e.g., 0.3 for recall optimization)
threshold = 0.3
y_pred_adjusted = np.where(y_probs[:, 1] >= threshold, 1, 0)

# Compute Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_probs[:, 1])
auprc = auc(recall, precision)

# Plot Precision-Recall Curve
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AUC = {auprc:.2f})')
plt.show()

# Generate classification report with adjusted predictions
report = classification_report(y_test, y_pred_adjusted, target_names=["No Diabetes (0)", "Diabetes/Risk (1)"])
print("Performance Metrics:")
print(report)

```
